```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import malv
pret = pd.read_csv(r'~/Desktop/Grain sizing data/Sedi_Test/Malvern_GS_Data/20200310.csv')
#before import remove % from headers.
#import csv file, change header number to get column headings as first row, for Retsch output = 24
r044=pd.read_csv(r'~/Desktop/Grain sizing data/Sedi_Test/0-44_X_AREA.csv', header=24)
print(r044)
```

```python
#find midpoint of class
r044['midpoint']= (((r044['Size class'] + r044['[mm]'])/2)*1000)
spht3 = r044['SPHT3']

#find boundaries from spreadsheet and then calc midpoint and select data
bound1 = pret.iloc[1, 9:66]
bound2 = pret.iloc[2, 8:65]
midpoint = (bound1 + bound2)/2
data044av = pret.iloc[5, 69:127]
data0442av = pret.iloc[8, 69:127]
```

```python
#plot percent in class from Retsch and Malvern against each other, limiting x,y axis range so comparable
#define variables in this box
midm = midpoint
midr = r044['midpoint']
m044 = data044av
r044 = r044['p3 ']
spher = spht3

fig1=plt.figure()
ax=fig1.add_subplot(111, label='1')
ax2=fig1.add_subplot(111,label='2', frame_on=False)
ax3=fig1.add_subplot(111,label='3', frame_on=False)
ax.plot(midm, m044, color='C0')
#set limits here
ax.set_ylim(0, 20)
ax2.set_ylim(0, 20)
ax.set_xlim(0, 500)
ax2.set_xlim(0, 500)
ax3.set_xlim(0,500)
ax.set_xlabel('Midpoint of class in micron')
ax.set_ylabel('Percentage in class')
ax.tick_params(axis='x', colors='C0')
ax.tick_params(axis='y', colors='C0')

ax2.plot(midr, r044, color='C1')


ax3.plot(midr, spher, color='C2')
ax3.xaxis.tick_top()
ax3.yaxis.tick_right()
ax3.set_ylabel('Sphericity')

ax.tick_params(axis='y', colors='C2')
ax3.xaxis.set_label_position('top') 
ax3.yaxis.set_label_position('right')

#plt.savefig('retsch vs malv.jpeg') #unhash first part to save 
```
